## Compoud Interest calculation without user input 


```python
def compound_interest(principal, rate, time):
    interest = principal*(1+ (rate/12)/100) ** (12*time)
    print('The compound interest is ${}'.format(interest))

compound_interest(5000, 5, 5)
```

## Compound interest calculation with user input

```python
def compound_interest():
    p = int(input("Enter the principal amount: "))
    r = int(input("Enter rate of interest: "))
    t = int(input("Enter time in years: " ))
    interest = p*(1+ (r/12)/100) ** (12*t)
    print('The compound interest is ${}'.format(interest))

compound_interest()
```
